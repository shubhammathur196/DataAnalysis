## Data cleaning 

```python
# Check for null values 
df.isnull().sum()
```

```python
#filling the null Values with mean 
df['math score'].fillna(df['math score'].mean() ,inplace=True)
```

```python
# Checking if any null Values remain
df.isnull().sum()
```

```python
# Dropping the features which are not so important 
student = df.drop(['race/ethnicity','parental level of education'],axis=1)
```

```python
student.head()
```

 #Relationship Analysis

```python
co=student.corr()
```

```python
# Using Seaborn to plot a Heat map
sns.heatmap(co,xticklabels=co.columns,yticklabels=co.columns,annot=True)
```

```python
# Using Seaborn to plot a Pairplot to look at the relationship between features
sns.pairplot(student)
```

```python
# Plotting a scatterplot as other plots wasn't very helpful
sns.relplot(x='math score',y='reading score',hue='gender',data=student)
```

```python
#Scatter plot 
sns.relplot(x='math score',y='reading score',hue='lunch',data=student)
```

```python
# Lets draw a histogram 
student.hist(bins=5)
```
